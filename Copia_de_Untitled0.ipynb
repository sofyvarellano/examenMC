```python
#Sofia Valeria Arellano Torres 2041735
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Leer el archivo CSV desde la URL y guardarlo en un DataFrame
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)

# Imprimir las últimas 5 filas y el resumen estadístico
print("--- Últimas 5 filas del dataset ---")
display(df.tail()) # Usamos display en Colab para que la tabla se vea con buen formato

print("\n--- Resumen estadístico descriptivo ---")
display(df.describe())

# Encontrar y eliminar filas donde x, y, o z sean exactamente 0
df = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]
print(f"\nTotal de filas después de limpiar las dimensiones nulas: {len(df)}")

# Crear nueva columna 'volumen_aprox' (x * y * z)
df['volumen_aprox'] = df['x'] * df['y'] * df['z']
```

```python
#Sofia Valeria Arellano Torres 2041735
# Gráfico de Caja de la columna
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'])
plt.title('Distribución Estadística del Precio (Boxplot)')
plt.ylabel('Precio (USD)')
plt.xticks([1], ['Precio']) # Limpia la etiqueta del eje X
plt.show()

# Gráfico de Dispersión
plt.figure(figsize=(8, 6))
# Se usa un poco de transparencia para ver mejor la densidad de puntos
plt.scatter(df['volumen_aprox'], df['price'], alpha=0.4, color='teal')
plt.title('Relación entre el Volumen Aproximado y el Precio')
plt.xlabel('Volumen Aproximado')
plt.ylabel('Precio (USD)')
plt.show()


plt.figure(figsize=(8, 6))

conteo_cortes = df['cut'].value_counts()

plt.bar(conteo_cortes.index, conteo_cortes.values, color=['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974'])
plt.title('Cantidad de Piezas por Tipo de Corte')
plt.xlabel('Tipo de Corte (Cut)')
plt.ylabel('Cantidad de Piezas')
plt.show()
```
